```python
import pandas as pd
import numpy as np
import json
import urllib.request,urllib.parse,urllib.error
from dotenv import load_dotenv
import os

load_dotenv()
#environment variables
auth_token = os.getenv("API_KEY")

url = f"https://api.themoviedb.org/3/movie/popular?api_key={auth_token}&language=en-US&page=1"


fhand = urllib.request.urlopen(url)

data = fhand.read().decode()

js= json.loads(data)
with open('data.json', 'w') as f:
  json.dump(js['results'], f, ensure_ascii=False)

dataz = pd.read_json('data.json',encoding='cp1252')

dataz = dataz.drop(['adult','original_language','backdrop_path','poster_path'],axis=1)
dataz.to_csv('popular_movie.csv')
dataz
```
